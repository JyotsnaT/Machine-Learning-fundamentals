```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def f(x, y):
  return x**2 + 2*y**2

def grad_f(x, y):
  return np.array([2*x, 4*y])

x = np.arange(-2, 2.1, 0.1)
y = np.arange(-2, 2.1, 0.1)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
Gx, Gy = grad_f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.quiver(X, Y, Z, Gx, Gy, length=0.01, color='b', W=0.001)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Gradient of f(x, y)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# Make the direction data for the arrows
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```

```python
def check_and_rotate(char, min_val, max_val, rotation_factor):
  if char >= min_val and char <= max_val:
      rel_pos = char - min_val
      range = max_val - min_val + 1

      rot_rel_pos = rel_pos + rotation_factor
      if rot_rel_pos < 0:
        rot_rel_pos += range-1
      elif rot_rel_pos >= range:
        rot_rel_pos = rot_rel_pos % range
      rot_char = rot_rel_pos + min_val
  else:
    return char
  return rot_char

#  char = 65
# rel_pos = 0
# rotation_factor = 200
# min_val = 65, max_val = 90, range = 26
# rot_rel_pos = 200
# rot_rel_pos = 18
# rot_char = 83


def rotationalCipher(input_str, rotation_factor):
  # Write your code here
  new_str = ""
  for char in input_str:
    print(char)
    char_num = ord(char)
    rot_char = check_and_rotate(char_num, ord('0'), ord('9'), rotation_factor)
    rot_char = check_and_rotate(rot_char, ord('A'), ord('Z'), rotation_factor)
    rot_char = check_and_rotate(rot_char, ord('a'), ord('z'), rotation_factor)
    print(rot_char, chr(rot_char))
    new_str += str(chr(rot_char))
    print(str(new_str))

  return new_str
```
